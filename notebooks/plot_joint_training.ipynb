```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mdrnn_vae_training = pd.read_csv('../exp_dir/joint/logger.txt', sep=' ')
mdrnn_vae_training.reset_index(inplace=True)
mdrnn_vae_training
```

```python
start_point = 0
end = 19

mdrnn_vae_training = mdrnn_vae_training.iloc[start_point:end]
```

```python


mdrnn_vae_training.plot('index', ['train_loss'])
plt.title('Summary losses')
#plt.ylim( -1e2, 1e3)
```

```python
mdrnn_vae_training.plot('index', ['test_loss'])
plt.title('Summary losses')
```

```python
mdrnn_vae_training.plot('index', ['test_max_reward_planner','test_avg_reward_planner' , 'test_min_reward_planner', 'test_std_reward_planner'])
plt.title('Rewards from Best Planner on Each Rollout')
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
for val in ['test_max_feef_planner','test_avg_feef_planner' , 'test_min_feef_planner', 'test_std_feef_planner']:
    plt.figure()
    mdrnn_vae_training.plot('index', val)
    plt.show()
```

```python
mdrnn_vae_training.plot('index', ['test_max_feef_planner','test_avg_feef_planner' , 'test_min_feef_planner', 'test_std_feef_planner'])
plt.title('FEEFs from Best Planner on Each Rollout')
plt.ylim(mdrnn_vae_training['test_min_feef_planner'].quantile(0.5), mdrnn_vae_training['test_min_feef_planner'].quantile(1.0))
```
