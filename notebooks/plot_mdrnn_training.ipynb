```python
import json
import numpy as np 
import matplotlib.pyplot as plt

loss_dict = json.load(open('../exp_dir/mdrnn/logger.json', 'r'))
```

```python
loss_dict.keys()
```

```python
plt.plot(loss_dict['train_loss'], label='train')
plt.plot(loss_dict['test_loss'], label='test')
plt.legend()
plt.ylim([-2, 200])
plt.ylabel('loss')
plt.xlabel('epochs')
```

```python
plt.plot(loss_dict['train_mse'], label='train')
plt.plot(loss_dict['test_mse'], label='test')
plt.legend()
plt.ylim([-2, 100])
plt.ylabel('mse')
plt.xlabel('epochs')
```

```python
plt.plot(loss_dict['train_gmm'], label='train')
plt.plot(loss_dict['test_gmm'], label='test')
plt.legend()
plt.ylim([-2, 10])
plt.ylabel('mse')
plt.xlabel('epochs')
```
